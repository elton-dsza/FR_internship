### <b>Question 2</b>

Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of
release).

```python
r = requests.get('http://www.imdb.com/chart/top')
soup = BeautifulSoup(r.text, 'lxml')

movies = soup.select('td.titleColumn')
links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]

imdb = []
for i in range(0, 99):
    movie_string = movies[i].get_text()
    movie_title = " ".join(movie_string.split()[1:-1])
    year = " ".join(movie_string.split()[-1:])[1:-1]
    data = {"movie_title": movie_title,
            "year": year,
            "star_cast": crew[i],
            "rating": ratings[i],
            "link": links[i]}
    imdb.append(data)
```

```python
pd.DataFrame(imdb)
```

### <b>Question 4</b>

Write a python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
r = requests.get('https://bookpage.com/reviews?book_genre=fiction&page=1')
soup = BeautifulSoup(r.text, 'lxml')
books = soup.find_all("h4", {"class" : "italic"})
authors = soup.find_all("p", {"class" : "sans bold"})
genres = soup.find_all("p", {"class" : "genre-links hidden-phone"})
reviews = soup.find_all("p", {"class" : "excerpt"})

book_list = []
for i in range(0,5):
    data = authors[i].get_text()
    author = " ".join(data.split())
    data = books[i].get_text()
    book = " ".join(data.split())
    data = genres[0].get_text()
    genre = " ".join(data.split())
    data = reviews[0].get_text()
    review = " ".join(data.split())
    dic = {"Books" : book,
          "Author" : author,
           "Genre" : genre,
           "Review" : review
          }
    book_list.append(dic)
```

```python
pd.DataFrame(book_list)
```

### <b>Question 7</b><br>
Write a python program to scrape details of all the mobile phones under Rs. 20,000 listed on Amazon.in. The
scraped data should include Product Name, Price, Image URL and Average Rating.

```python
HEADERS = ({'User-Agent':
            'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
            'Accept-Language': 'en-US, en;q=0.5'})


URL = "https://www.amazon.in/s?i=electronics&bbn=1389401031&rh=n%3A1389401031%2Cp_36%3A-2000000&s=price-desc-rank&dc&qid=1626954088&rnid=1318502031&ref=sr_st_price-desc-rank"
webpage = requests.get(URL, headers=HEADERS)
```

```python
soup = BeautifulSoup(webpage.content, "lxml")
```

```python
titles = soup.find_all("span", attrs={"class":'a-size-base-plus a-color-base a-text-normal'})
image_links = soup.find_all("img", attrs={"class":'s-image'})
average_ratings = soup.find_all("span", attrs={"class":'a-icon-alt'})
Prices = soup.find_all("span", attrs={"class":'a-offscreen'})
```

```python
info = []

for i in range(1,23):
    title = " ".join(titles[i].get_text().split())
    d = str(image_links[i])
    links = re.findall(r'(https?://\S+)', d)
    info.append({"Product Title":title, "Image Links": str(links[1]), "Price" : Prices[i].get_text(),"Average Rating":average_ratings[i].get_text()[:3]})
```

```python
pd.DataFrame(info)
```

### <b>Question 8</b><br>
Write a python program to extract information about the local weather from the National Weather Service
website of USA, https://www.weather.gov/ for the city, San Francisco. You need to extract data about 7 day
extended forecast display for the city. The data should include period, short description, temperature and
description.

```python
r = requests.get('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996')
soup = BeautifulSoup(r.text, 'lxml')
```

```python
periods = soup.find_all("p", {"class" : "period-name"})
short_descriptions = soup.find_all("p", {"class" : "short-desc"})
temp_high = soup.find_all("p", {"class" : "temp temp-high"})
temp_low = soup.find_all("p", {"class" : "temp temp-low"})
pos_a = 0
pos_b = 0
flag = 0
temp_ordered = []
for i in range(0,len(temp_high)+len(temp_low)):
    if flag == 0:
        temp_ordered.append(temp_high[pos_a].get_text())
        pos_a += 1
        flag = 1
    else :
        temp_ordered.append(temp_low[pos_b].get_text())
        pos_b += 1
        flag = 0
detailed_descriptions = soup.find_all("div", {"class" : "col-sm-10 forecast-text"})

info = []
for i in range(0,len(periods)):
    info.append({"Period" : periods[i].get_text(),
                 "Short Description" :short_descriptions[i].get_text(),
                 "Temperature" : temp_ordered[i] ,
                 "Description" : detailed_descriptions[i].get_text()
                })
    
```

```python
pd.DataFrame(info)
```
